# 2. Exploratory data analysis with DESeq2
The number of reads mapped to a gene cannot be corelated directly to gene expression levels. Technical and biological factors as amplification efficiency, sequencing depth, gene length or expresion levels can strongly influence read counts. Thus, it is necessary to quantify the number of reads assigned to each gene relative to the total number of reads and total RNA.



```r
# get the table of read counts by indicating the path to the file
readcounts <-read.table ("./raw_countstdl.tsv", header = TRUE)

head(readcounts)
str(readcounts)
```

```r
 # make a data.frame with meta-data 
 sample_info <- data.frame(condition = gsub ( "[1-4]+" , "" , names(readcounts)), row.names = names(readcounts))
 sample_info
```

```r
# generate the DESeqDataSet
DESeq.ds <- DESeqDataSetFromMatrix(countData = readcounts, colData = sample_info, design = ~condition )

# check the result:
colData(DESeq.ds) %>% head
assay(DESeq.ds, "counts") %>% head #same as counts(DESeq.ds)
```

```r
# test what counts() returns
counts(DESeq.ds) %>% str
options(repr.plot.width=5, repr.plot.height=5)
# remove genes without any counts
DESeq.ds <- DESeq.ds [rowSums(counts(DESeq.ds)) > 0, ]

# investigate different library sizes
colSums(counts(DESeq.ds)) == colSums(readcounts) # should be the same as colSums (readcounts)
```

## Normalization
DESeq2's normalization function estimateSizeFactors() follows three steps:
* for every gene (row), the geometric mean of read counts across all samples (ctl1-4 and treat1-4) is calculated; this yields the "pseudo-reference" - one value per row;
* every read value in the table is divided by its correspnding pseudo-reference value;
* for every sample (column), the size factor is the median of ratios from step 2. 

```r
# calculate the size factor and add it to the data set
DESeq.ds <- estimateSizeFactors(DESeq.ds)
sizeFactors(DESeq.ds)
options(repr.plot.width=5, repr.plot.height=5)
# if you check colData () again , you see that this now contains the sizeFactors
colData(DESeq.ds)

# counts() allows you to immediately retrieve the normalized read counts
counts.sf_normalized <- counts(DESeq.ds, normalized = TRUE)
```

## Transformation
Due to the wide range of RNA expression values, data is log-transformed for downstream analysis (most of the times log2).

```r
# transform size-factor normalized read counts to log2 scale using pseudocount of 1 (+1 because log2(0))
log.norm.counts <- log2(counts.sf_normalized + 1)

# plot layout
layout(matrix(c(1, 2), nrow=1, byrow=TRUE),)
options(repr.plot.width=8, repr.plot.height=5)
# boxplots of non-transformed read counts ( one per sample )
boxplot(counts.sf_normalized, notch = TRUE , main = "untransformed read counts" , ylab = "read counts", las = 2)

# box plots of log2 - transformed read counts
boxplot(log.norm.counts, notch = TRUE , main = "log2-transformed read counts" , ylab = "log2(read counts)", las = 2)
```

Many statistical tests assume that all variables in a data set have similar variance. This can be checked by plotting the mean vs. standard deviation. In case large differences in variance can be observed, DESeq2 provides a different type of transformation including variance shrinkage via the rlog() function. rlog() output is normalized and log2 transformed where values are adjusted to fit the general variance-mean trend. 


```r

options(repr.plot.width=4, repr.plot.height=4)
#mean-sd plot normalized log2-transformed data
msd_plot <- meanSdPlot(log.norm.counts, ranks = FALSE, plot = FALSE )
msd_plot$gg+
    ggtitle("normalized log2(read counts)") +
    ylab("standard deviation")

#set blind=FALSE if the different conditions lead to strong differences in a large proportion of the genes. 
DESeq.rlog <- rlog(DESeq.ds, blind = TRUE)
rlog.norm.counts <- assay(DESeq.rlog)

#mean-sd plot for rlog-transformed data
msd_plot2 <- meanSdPlot(rlog.norm.counts, ranks = FALSE, plot = FALSE)
msd_plot2$gg +
    ggtitle("rlog-transformed read counts") +
    ylab("standard deviation") +
    coord_cartesian(xlim = c(0, 20), ylim = c(0, 5))

```

## Hierarchical clustering

Hierarchical clustering can be used to evaluate the degree of similarity between different samples. This method is based on pairwise comparisons of individual samples (columns) which are grouped into "branches" based on a matrix of similarity metrics. Here we use the Pearson correlation coefficient r to determine the distance between samples. As correlation between samples increases, the r value approaches 1. Distance d is calculated as 1 - r, so a high correlation coefficient translates as a small distance measure and viceversa and, conversely, a low r to a large distance.

```r
#A dendrogram can be generated in R using the functions cor(), as.dist(), and hclust():

# cor () calculates the correlation between columns of a matrix
distance.m_rlog <- as.dist(1 - cor(rlog.norm.counts , method = "pearson"))

# plot () can directly interpret the output of hclust ()
options(repr.plot.width=5, repr.plot.height=6)o
plot(hclust(distance.m_rlog), xlab = "", ##hclust() default is complete - largest distance between any 2 members
    labels = colnames(rlog.norm.counts) ,
    main = "rlog transformed read counts\ndistance: Pearson correlation")
```

## Principal component analysis

# 3. Differential Gene Expression (DGE) Analysis

Compare the effect of the treatment (treat) versus the control(ctl) samples, with the control values used as the denominator for the fold change calculation.
## 3.1 DESeq2


```r
# DESeq2 uses the levels of the condition to determine the order of the comparison
str(colData(DESeq.ds)$condition )

# set ctl as the first-level-factor
colData(DESeq.ds)$condition <- relevel(colData(DESeq.ds)$condition, "ctl")

# run DGE analysis
DESeq.ds <- DESeq(DESeq.ds)
```

The results() function lets you extract the base means across samples, moderated log2 fold changes, standard errors, test statistics etc. for every gene.


```r
DGE.results <- results(DESeq.ds, independentFiltering = TRUE, alpha = 0.05)
summary (DGE.results)

#the DESeqResult object can basically be handled like a data . frame
head(DGE.results)
table(DGE.results$padj < 0.05) #rownames(subset(DGE.results, padj < 0.05))
```

### Exploratory plots

"Once you’ve obtained a list of p-values for all the genes of your data set, it is important to realize that you
just performed the same type of test for thousands and thousands of genes. That means, that even if you
decide to focus on genes with a p-value smaller than 0.05, if you’ve looked at 10,000 genes your final list may
contain 0.05 ∗ 10, 000 = 500 false positive hits. To guard yourself against this, all the tools will offer some sort
of correction for the multiple hypotheses you tested, e.g. in the form of the Benjamini-Hochberg formula. You
should definitely rely on the adjusted p-values rather than the original ones to zoom into possible candidates
for downstream analyses and follow-up studies."


```r
options(repr.plot.width=8, repr.plot.height=8)
par(mfrow = c(2,2))
hist(DGE.results$pvalue,
    xlim=c(0.0,1.0),
    ylim = c(0,4000),
    col = "grey", border = "white", xlab = "p-value", ylab = "count",
    main = "frequencies of p-values")
hist(DGE.results$padj,
    xlim=c(0.0,1.0),
    ylim = c(0,4000),
    col = "grey", border = "white", xlab = "p-value", ylab = "count",
    main = "frequencies of adjusted p-values")

plotMA(DGE.results, alpha = 0.05, main = "control vs. treatment",
    ylim = c(-8,8))
```

```r
# sort the results according to the adjusted p - value
DGE.results.sorted <- DGE.results[order(DGE.results$padj),]

#select genes with p-value<0.05
DGE.results.sorted <- subset(DGE.results.sorted, padj < 0.05)

#order by fold change(absolute value)
#DGE.results.sorted.byfold <- DGE.results.sorted[order(abs(DGE.results.sorted$log2FoldChange), decreasing = TRUE),]
#DGE.results.sorted.byfold$log2FoldChange
#DGEgenes <- rownames(subset(DGE.results.sorted.byfold, abs(log2FoldChange) >= 1))

#identify genes with the desired adjusted p-value cut-off
DGEgenes <- rownames(DGE.results.sorted)
length(DGEgenes)
```

## 3.2 edgeR

```r
sample_info.edger <- factor(c(rep("ctl", 4) , rep("treat" , 4)))
sample_info.edger <- relevel(sample_info.edger, ref = "ctl" )
edgeR.DGElist <- DGEList(counts = readcounts, group = sample_info.edger)
head(edgeR.DGElist$counts)
edgeR.DGElist$samples
```

```r
#edgeR also recommends removing genes with almost no coverage. In order to determine a sensible cutoff, we
#plot a histogram of counts per million calculated by edgeR’s cpm() function.

options(repr.plot.width=4, repr.plot.height=4)
#get an impression of the coverage across samples
hist(log2(rowSums(cpm(edgeR.DGElist))), main = "log2 cpm/gene")
summary(log2(rowSums(cpm(edgeR.DGElist))))
```

```r
#remove genes that do not have one count per million in at least 5 samples 

keep <- rowSums(cpm(edgeR.DGElist)>=1)>= 4
edgeR.DGElist <- edgeR.DGElist[keep ,]
```

```r
# recompute library sizes after filtering
edgeR.DGElist$samples$lib.size <- colSums(edgeR.DGElist$counts)
head(edgeR.DGElist$samples)

 
```

```r
#Calculate normalization factors for the library sizes. We use the standard edgeR method here, which is
#the trimmed mean of M-values; if you wanted to use, for example, DESeq’s size factor, you could use
#method = "RLE")
edgeR.DGElist <- calcNormFactors(edgeR.DGElist, method = "TMM")
edgeR.DGElist$samples
```

```r
 #specify the design setup  [ ~ sample _ info . edger ] you can see that it ' s
 #exactly what we used for DESeq2 , too
design <- model.matrix(~sample_info.edger)

# estimate the dispersion for all read counts across all samplesoptions(repr.plot.width=5, repr.plot.height=5)
edgeR.DGElist <- estimateDisp(edgeR.DGElist, design)
 
# fit the negative binomial model
edger_fit <- glmFit(edgeR.DGElist, design)

# perform the testing for every gene using the neg. binomial model
edger_lrt <- glmLRT(edger_fit)
 
```

```r
 # extract results from edger_lrt$ table plus adjusted p-values
DGE.results_edgeR <- topTags(edger_lrt, n = Inf , 
    sort.by = "PValue", adjust.method = "BH")
head(DGE.results_edgeR$table)


DGEgenes.edgeR <- rownames(subset(DGE.results_edgeR$table, FDR < 0.05))
str(DGE.results_edgeR$table[DGEgenes.edgeR,])

```

## 3.3 limma-voom

```r
library(limma)
# use edgeR to remove lowly expressed genes and normalize reads for sequencing depth as above
edgeR.DGElist <- DGEList(counts = readcounts, group=sample_info.edger)
keep <- rowSums(cpm(edgeR.DGElist) >= 1) >= 4
edgeR.DGElist <- edgeR.DGElist[keep, ]
edgeR.DGElist <- calcNormFactors(edgeR.DGElist, method = "TMM")

#design matrix
design <- model.matrix(~sample_info.edger)

# transform the count data to log2-counts-per-million and estimate the mean - variance relationship , which is used
#to compute weights for each count. The weights can be fit to a lm
rownames(design) <- colnames(edgeR.DGElist)
voomTransformed <- voom(edgeR.DGElist, design, plot = FALSE)
```

```r
# fit a linear model for each gene
voomed.fitted <- lmFit(voomTransformed, design = design)

# compute moderated t - statistics , moderated F - statistics , and log - odds of differential expression
voomed.fitted <- eBayes(voomed.fitted)
```

```r
colnames(design) # check how the coefficient is named
```

```r
# extract gene list with logFC and statistical measures
DGE.results_limma <- topTable(voomed.fitted, coef = "sample_info.edgertreat",
 number = Inf, adjust.method = "BH", sort.by = "p" )

str(DGE.results_limma)
head(DGE.results_limma)
```

```r
DGEgenes.limma <- rownames(subset(DGE.results_limma, adj.P.Val < 0.05))
str(DGEgenes.limma)
```

# 4. Interpretation of results
## Venn diagram
A Venn diagram can be used to check agreement between the three methods. (adjusted p-values < 0.05 for deseq and limma, FDR(false discovery rate) < 0.05 for edger)

```r
DE_list <- list(edger = rownames(subset(DGE.results_edgeR$table, FDR <=0.05)),
     deseq2 = rownames(subset(DGE.results, padj <=0.05)),
     limma = rownames (subset(DGE.results_limma,adj.P.Val <=0.05)))
options(repr.plot.width=5, repr.plot.height=5)
venn<- gplots::venn(DE_list)

#add all common DE gene names into one vector
DGEgenes.intersection <- attr(venn, "intersections")$`edger:deseq2:limma`
DGEgenes.intersection2 <- attr(venn, "intersections")$`edger:deseq2`
```

## Same thing but for college students

```r
# fancier venn alternative 
options(repr.plot.width=8, repr.plot.height=5)
DE_gns <- UpSetR::fromList(DE_list)
UpSetR::upset(DE_gns, order.by = "freq")
```

## Correlation of log fold change 

```r
#correlation of logFC for genes found DE in all three tools
DE_gns_all <- intersect(intersect(DGEgenes,DGEgenes.limma),DGEgenes.edgeR)

#make a data . frame of fold change values
 DE_fc <- data.frame( edger = DGE.results_edgeR[DE_gns_all, ]$table$logFC ,
         limma = DGE.results_limma[DE_gns_all,]$logFC,
         deseq2 = DGE.results[DE_gns_all,]$log2FoldChange ,
         row.names = DE_gns_all)
```

```r
# visually check how well the estimated logFC of the different tools agree for the DE genes
options(repr.plot.width=5, repr.plot.height=5)
ggpairs(DE_fc)
```
